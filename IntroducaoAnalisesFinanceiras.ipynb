# Carregando Arquivos

```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
#Vamos realizar a pesquisa com as ações da Apple
raw = yf.download ('SPY AAPL', start = '2010-01-01', end='2025-04-08', auto_adjust=False)
```

```python
raw
```

```python
raw.columns
```

```python
#documentação do método pipe
raw.pipe?
```

```python
#Vamos retirar os nomes empilhados e pegar só os nomes de cima, jogar fora os de baixo e escolher as colunas de duas em duas.
def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns]
    df.columns = outer
    return df
    
(raw.iloc[:, ::2].pipe(fix_cols))
    
```

# Gráfico de Velas

```python
#Ano de 2025
fig, ax = plt.subplots(figsize=(14,5))

def plot_candle(df, ax):
    # Vela
    ax.vlines(x=df.index, ymin=df.Low, ymax=df.High, colors ='k', linewidth=1)
    # Vermelho - Decréscimo
    red = df.query('Open > Close')
    ax.vlines(x=red.index, ymin=red.Close, ymax=red.Open, colors='r', linewidth=3)
    # Verde - Acréscimo
    green = df.query('Open <= Close')
    ax.vlines(x=green.index, ymin=green.Close, ymax=green.Open, colors='g', linewidth=3)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return df

(raw.iloc[:,::2].pipe(fix_cols).resample('d').agg({'Open':'first', 'High':'max', 'Low':'min', 'Close':'last'}).loc['jan 2025': 'apr 2025']
 .pipe(plot_candle, ax))
```

```python
#Vamos verificar os ultimos 100 fechamentos !!! ERROO no eixo X  - Pandas Categorizou !!! 
fig, ax = plt.subplots(figsize=(15,4))
(aapl.pct_change().Close.iloc[-100:].plot.bar(ax = ax))
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
ax.xaxis.set_minor_locator(dates.DayLocator())
```

```python
def my_bar(ser, ax):
    ax.bar(ser.index,ser)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return ser

fig, ax = plt.subplots(figsize=(14,4))
(aapl.pct_change().Close.iloc[-100:].pipe(my_bar, ax))
```
