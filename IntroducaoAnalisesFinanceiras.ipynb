# Carregando Arquivos

```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
#Vamos realizar a pesquisa com as ações da Apple
raw = yf.download ('SPY AAPL', start = '2010-01-01', end='2025-04-08', auto_adjust=False)
```

```python
#documentação do método pipe
raw.pipe?
```

```python
#Vamos retirar os nomes empilhados e pegar só os nomes de cima, jogar fora os de baixo e escolher as colunas de duas em duas.
def fix_cols(df):
    columns = df.columns
    outer = [col[0] for col in columns]
    df.columns = outer
    return df
    
(raw.iloc[:, ::2].pipe(fix_cols))
    
```

# Plots

- Plots em Linha
- Matplotlib plots


```python
(raw.iloc[:,:-2:2].pipe(fix_cols).plot())
```

```python
(raw.iloc[:,::2].pipe(fix_cols).Close.plot())
```

```python
(raw.iloc[:,::2].pipe(fix_cols).Volume.plot())
```
